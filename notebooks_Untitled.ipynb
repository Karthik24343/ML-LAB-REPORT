```python
import numpy as np
import pandas as pd

# Task A

data = {
    "RollNo":[101,102,103,104,105,106,107,108,109,110],
    "Name":["Ravi","Priya","Kiran","Anu","Sita","Raj","Divya","Arun","Keerthi","Naveen"],
    "Gender":["M","F","M","F","F","M","F","M","F","M"],
    "Attendance":[85,92,78,95,88,70,90,82,96,75],
    "InternalMarks":[32,38,28,40,35,25,37,30,39,27],
    "ExternalMarks":[55,58,45,60,54,40,57,48,59,44]
}

df = pd.DataFrame(data)

print(df)

print("First 5 records: \n",df.head())
print("Last 5 records: \n",df.tail())

rows,columns = df.shape
print("No of rows: ",rows)
print("No of columns: ",columns)

print(df.columns.tolist())

print(df.dtypes)
print("Dataset Information: ")
df.info()

# Task B
print("Mean Attendance: ",df["Attendance"].mean())
print("Average Internal Marks: ",df["InternalMarks"].mean())
print("Average External Marks: ",df["ExternalMarks"].mean())
print("Maximum Internal Marks: ",df["InternalMarks"].max())
print("Minimum Internal Marks: ",df["InternalMarks"].min())
print("Maximum External Marks: ",df["ExternalMarks"].max())
print("Minimum External Marks: ",df["ExternalMarks"].min())
print("Standard Deviation of Internal Marks: ",df["InternalMarks"].std())
print("Summary Statistics: \n",df.describe())

# Task C
df["TotalMarks"] = df["InternalMarks"] + df["ExternalMarks"]
df["Percentage"] = df["TotalMarks"]

print("Updated DataFrame: \n",df)

# Task D
print("The details of students whose attendance is greater than 90%: \n",df[df["Attendance"] > 90])
print("Details of students whose TotalMarks exceed 90: \n",df[df["TotalMarks"] > 90])
print("Details of female students: \n",df[df["Gender"]=='F'])
print("Details of male students: \n",df[df["Gender"]=='M'])
print("Students whose ExternalMarks are less than 50: \n",df[df["ExternalMarks"] < 50])

# Task E
print("Student with highest TotalMarks \n",df.loc[df["TotalMarks"].idxmax()])
print("Student with lowest TotalMarks \n",df.loc[df["TotalMarks"].idxmin()])

print("Top 3 students: \n",
      df.sort_values(by="TotalMarks", ascending=False).head(3))

print("Bottom 3 students: \n",
      df.sort_values(by="TotalMarks", ascending=True).head(3))

average_marks = df["TotalMarks"].mean()

count = len(df[df["TotalMarks"] > average_marks])
print("Number of students scoring above class average:", count)

# Task F
male_students = df[df["Gender"] == "M"]
averagem = male_students["TotalMarks"].mean()
print("Average TotalMarks of Male Students:", averagem)

female_students = df[df["Gender"] == "F"]
averagef = female_students["TotalMarks"].mean()
print("Average TotalMarks of Female Students:", averagef)

male_avg = df[df["Gender"] == "M"]["Attendance"].mean()
female_avg = df[df["Gender"] == "F"]["Attendance"].mean()

print("Average Attendance of Male Students:", male_avg)
print("Average Attendance of Female Students:", female_avg)

# Task G
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["TotalMarks"])
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.title("Total Marks of Students")
plt.xticks(rotation=30)
plt.show()

plt.figure(figsize=(8,5))
plt.hist(df["Attendance"], bins=10)
plt.xlabel("Attendance")
plt.ylabel("Number of Students")
plt.title("Histogram of Attendance")
plt.show()

# Pie Chart
gender_count = df["Gender"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_count,
        labels=gender_count.index,
        autopct='%1.1f%%')

plt.title("Distribution of Male and Female Students")
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(df["Attendance"], df["TotalMarks"])
plt.xlabel("Attendance")
plt.ylabel("Total Marks")
plt.title("Attendance vs Total Marks")
plt.show()

print("Relationship between Attendance and TotalMarks: \n",
      df["Attendance"].corr(df["TotalMarks"]))
```
